```python
import numpy as np
import pandas as pd
```

```python
#There is a blank line as first row in csv file
football_df= pd.read_csv("Laliga.csv", skip_blank_lines=True,skiprows=1)#skip rows 
```

```python
#Q1 Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (2.5 points)
# We are replacing with string 0 as we are going to invoke to_numeric in next steps
football_df.replace(to_replace='-', value='0', inplace=True)
#We will perform to numeric on columns other than the columns which are either already numeric or which are of the type which
#doesnt need to be converted to numeric
cols = football_df.columns.drop(["Pos", "Team", "Seasons", "Debut", "Since/LastApp","BestPosition"])
football_df[cols] = football_df[cols].apply(pd.to_numeric, errors='coerce')
football_df.dtypes
```

```python
football_df.tail().T
```

```python
#Q2Print all the teams which have started playing between 1930-1980. (5 points)
#Q2Step1 - Since debut is a string with hyphen so we cannot do operations on it so split the debut into two 
# columns start and end
football_df[['DebutStart','DebutEnd']] = football_df['Debut'].str.split('-', expand=True)
#prepend first two characters of year like 19 or 20 to debut end in all dates except where debut end is None
football_df.loc[(football_df['DebutEnd'].isnull() == False),['DebutEnd']] = football_df['DebutStart'].str[:2]+football_df['DebutEnd'].astype(str)
football_df[['DebutStart','DebutEnd']]
```

```python
#Q2 Step2 Remove the debut column as it has been split into two columns
football_df.drop("Debut", inplace=True, axis=1)
#remove None values from Debut End
football_df.loc[football_df["DebutEnd"].isnull() == True,"DebutEnd"] = '0'
football_df[['DebutStart','DebutEnd']]
```

```python
#Q2 Step 3 convert all debut columns to numeric 
football_df["DebutStart"] = football_df["DebutStart"].apply(pd.to_numeric, errors = 'coerce')
football_df["DebutEnd"] = football_df["DebutEnd"].apply(pd.to_numeric, errors = 'coerce')

#Q2 Step 4 Get all teams with debut between 1930-1980
teams = (football_df[((football_df['DebutStart'] >= 1930) & (football_df['DebutStart'] <= 1980)) |
           ((football_df['DebutEnd'] >= 1930) & (football_df['DebutEnd'] <= 1980))][['Team','DebutStart','DebutEnd']])
print(teams)
teams.info()
```

```python
#Q3 Print the list of teams which came Top 5 in terms of points (2.5 points)
#Laliga.csv pos column indicates position of team and Points indicates points held by each team
# pos column is already in ascending order and corresponding points are in descending order so top 5 rows represent team with 
# top 5 points
#football_df.loc[0:4, ['Pos', 'Team', 'Points']]
#To ensure that if csv changes still we get desired result we will use nlargest
(football_df.nlargest(5, 'Points'))[['Pos','Team','Points']]
```

```python
#Q4 Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences.
#Using the same function, find the team which has maximum and minimum goal difference. (5 points)
#Goal_diff_count = GoalsFor - GoalsAgainst
def Goal_diff_count():
      return football_df['GoalsFor'] - football_df['GoalsAgainst']
        
goaldiff_df = Goal_diff_count()
print("The team with maximum goal difference is\n",football_df.loc[goaldiff_df.idxmax(),["Team","GoalsFor", "GoalsAgainst"]]) 
print("The team with minimum goal difference is\n",football_df.loc[goaldiff_df.idxmin(),["Team","GoalsFor", "GoalsAgainst"]]) 
```

```python
#Q5
#Q5 1Create a new column with name “Winning Percent” and append it to the data set (5 points)
# Percentage of Winning = (GamesWon / GamesPlayed)*100
football_df["Winning Percent"] = (football_df["GamesWon"]/football_df["GamesPlayed"])*100 
```

```python
#Q5 2 If there are any numerical error, replace it with 0%
print("Before :\n", football_df[["Winning Percent"]])
football_df[football_df["Winning Percent"].isna()] = 0
print("After :\n",football_df[["Winning Percent"]])
```

```python
#Q5 3 Print the top 5 teams which has the highest Winning percentage
(football_df.nlargest(5, "Winning Percent")).T
```

```python
#Q6 Group teams based on their “Best position” and print the sum of their points for all positions (5 points)
#Eg: Best Position                Points
#        1                              25000
#        2                              7000for
(football_df.groupby(["BestPosition"]))[['Points']].sum()
```
